```python
"""
–û—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏:
- –†–∞–±–æ—Ç–∞ —Å —Ä–µ–∞–ª—å–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ captured_data.bin
- –ó–∞—â–∏—Ç–∞ –æ—Ç –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏—è –∏ infinity –∑–Ω–∞—á–µ–Ω–∏–π
- –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏–µ –∞–Ω–æ–º–∞–ª–∏–π
- –ü–æ–ª–Ω–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
- –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –º–∏—Ä–æ–≤—ã–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∞–º –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è
"""

import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Dict, Any
import logging
import warnings
import os
warnings.filterwarnings('ignore')

class PipelineRealDataSystemFixed:
    """
    –ü–û–õ–ù–ê–Ø —Å–∏—Å—Ç–µ–º–∞ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏ —Ç—Ä—É–±–æ–ø—Ä–æ–≤–æ–¥–æ–≤ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    
    –í–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏:
    - –ó–∞–≥—Ä—É–∑–∫–∞ –∏ –ø–∞—Ä—Å–∏–Ω–≥ –±–∏–Ω–∞—Ä–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    - –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏—á–µ—Å–∫–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    - –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ Random Forest
    - –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏–µ –∞–Ω–æ–º–∞–ª–∏–π
    - –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    """
    
    def __init__(self, random_state: int = 42):
        """–ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Å–∏—Å—Ç–µ–º—ã –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏"""
        self.random_state = random_state
        self.model = None
        self.scaler = StandardScaler()
        self.is_trained = False
        self.feature_names = None
        
        # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ª–æ–≥–∏—Ä–æ–≤–∞–Ω–∏—è
        logging.basicConfig(level=logging.INFO, 
                          format='%(levelname)s:%(name)s:%(message)s')
        self.logger = logging.getLogger(__name__)
    
    def load_binary_data(self, filepath: str) -> np.ndarray:
        """
        –ó–∞–≥—Ä—É–∂–∞–µ—Ç –∏ –ø–∞—Ä—Å–∏—Ç –±–∏–Ω–∞—Ä–Ω—ã–π —Ñ–∞–π–ª —Å –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º —Ñ–æ—Ä–º–∞—Ç–∞
        
        Args:
            filepath: –ø—É—Ç—å –∫ –±–∏–Ω–∞—Ä–Ω–æ–º—É —Ñ–∞–π–ª—É
            
        Returns:
            numpy array —Å –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã–º–∏ –∏ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
        """
        if not os.path.exists(filepath):
            raise FileNotFoundError(f"–§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω: {filepath}")
        
        file_size = os.path.getsize(filepath)
        self.logger.info(f"–†–∞–∑–º–µ—Ä —Ñ–∞–π–ª–∞: {file_size:,} –±–∞–π—Ç")
        
        try:
            # –ü—Ä–æ–±—É–µ–º —Ä–∞–∑–Ω—ã–µ —Ñ–æ—Ä–º–∞—Ç—ã –¥–∞–Ω–Ω—ã—Ö
            formats_to_try = [
                (np.float32, "32-bit float"),
                (np.float64, "64-bit float"),
                (np.int32, "32-bit integer"),
                (np.int16, "16-bit integer"),
                (np.uint8, "8-bit unsigned")
            ]
            
            for data_type, type_name in formats_to_try:
                try:
                    data = np.fromfile(filepath, dtype=data_type)
                    
                    if len(data) > 0:
                        self.logger.info(f"–î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã: {data.shape}, —Ç–∏–ø: {type_name}")
                        
                        # –ö–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –≤ float64 –¥–ª—è –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–π
                        data = data.astype(np.float64)
                        
                        # –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞ —ç–∫—Å—Ç—Ä–µ–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ –ø—Ä–∏–º–µ–Ω—è–µ–º –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—é
                        if np.any(np.abs(data) > 1e6):
                            self.logger.warning("–û–±–Ω–∞—Ä—É–∂–µ–Ω—ã –±–æ–ª—å—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è, –ø—Ä–∏–º–µ–Ω—è–µ–º –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—é")
                            max_abs_value = np.max(np.abs(data))
                            data = data / max_abs_value
                            self.logger.info(f"–î–∞–Ω–Ω—ã–µ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω—ã, –¥–∏–∞–ø–∞–∑–æ–Ω: [{np.min(data):.6f}, {np.max(data):.6f}]")
                        
                        return data
                        
                except Exception as e:
                    continue
                    
            raise ValueError("–ù–µ —É–¥–∞–ª–æ—Å—å –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Ñ–æ—Ä–º–∞—Ç –¥–∞–Ω–Ω—ã—Ö")
            
        except Exception as e:
            self.logger.error(f"–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ —Ñ–∞–π–ª–∞: {e}")
            raise
    
    def parse_sensor_data(self, raw_data: np.ndarray) -> pd.DataFrame:
        """
        –ü–∞—Ä—Å–∏—Ç —Å—ã—Ä—ã–µ –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç—Ä—É–∫—Ç—É—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç DataFrame
        
        Args:
            raw_data: –º–∞—Å—Å–∏–≤ —Å—ã—Ä—ã—Ö –¥–∞–Ω–Ω—ã—Ö
            
        Returns:
            DataFrame —Å–æ —Å—Ç—Ä—É–∫—Ç—É—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ —Å–µ–Ω—Å–æ—Ä–æ–≤
        """
        try:
            # –ü—Ä–æ–±—É–µ–º —Ä–∞–∑–ª–∏—á–Ω—ã–µ –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏–∏ –∫–∞–Ω–∞–ª–æ–≤
            possible_channels = [1, 2, 4, 8, 16, 32, 64]
            
            for n_channels in possible_channels:
                if len(raw_data) % n_channels == 0:
                    samples_per_channel = len(raw_data) // n_channels
                    
                    # Reshape –¥–∞–Ω–Ω—ã—Ö
                    structured_data = raw_data.reshape(samples_per_channel, n_channels)
                    
                    # –°–æ–∑–¥–∞–µ–º –æ—Å–º—ã—Å–ª–µ–Ω–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫
                    column_names = [f'sensor_{i+1}' for i in range(n_channels)]
                    df = pd.DataFrame(structured_data, columns=column_names)
                    
                    self.logger.info(f"–î–∞–Ω–Ω—ã–µ —Å—Ç—Ä—É–∫—Ç—É—Ä–∏—Ä–æ–≤–∞–Ω—ã: {df.shape}, –∫–∞–Ω–∞–ª–æ–≤: {n_channels}")
                    self.logger.info(f"–î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π: min={df.min().min():.6f}, max={df.max().max():.6f}")
                    self.logger.info(f"–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {df.mean().mean():.6f}, –°–ö–û: {df.std().mean():.6f}")
                    
                    return df
            
            # –ï—Å–ª–∏ –Ω–µ —É–¥–∞–ª–æ—Å—å –Ω–∞–π—Ç–∏ –ø–æ–¥—Ö–æ–¥—è—â–µ–µ —Ä–∞–∑–±–∏–µ–Ω–∏–µ, —Å–æ–∑–¥–∞–µ–º –æ–¥–Ω–æ–∫–∞–Ω–∞–ª—å–Ω—ã–π DataFrame
            df = pd.DataFrame({'sensor_1': raw_data})
            self.logger.info(f"–°–æ–∑–¥–∞–Ω –æ–¥–Ω–æ–∫–∞–Ω–∞–ª—å–Ω—ã–π DataFrame: {df.shape}")
            return df
            
        except Exception as e:
            self.logger.error(f"–û—à–∏–±–∫–∞ –ø–∞—Ä—Å–∏–Ω–≥–∞ –¥–∞–Ω–Ω—ã—Ö: {e}")
            raise
    
    def extract_features_safe(self, sensor_data: pd.DataFrame) -> pd.DataFrame:
        """
        –ë–ï–ó–û–ü–ê–°–ù–û–ï –∏–∑–≤–ª–µ—á–µ–Ω–∏–µ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏—á–µ—Å–∫–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å –∑–∞—â–∏—Ç–æ–π –æ—Ç –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏—è
        
        –ò–∑–≤–ª–µ–∫–∞–µ—Ç —Å–ª–µ–¥—É—é—â–∏–µ –≥—Ä—É–ø–ø—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:
        - –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ (—Å—Ä–µ–¥–Ω–µ–µ, –°–ö–û, RMS, –ø–∏–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ)
        - –§–æ—Ä–º-—Ñ–∞–∫—Ç–æ—Ä—ã (—ç–∫—Å—Ü–µ—Å—Å, –∞—Å–∏–º–º–µ—Ç—Ä–∏—è, –∫—Ä–µ—Å—Ç-—Ñ–∞–∫—Ç–æ—Ä)
        - –ß–∞—Å—Ç–æ—Ç–Ω—ã–µ (–ø–∏–∫–æ–≤–∞—è —á–∞—Å—Ç–æ—Ç–∞, —Å—Ä–µ–¥–Ω—è—è –∏ –°–ö–û –∞–º–ø–ª–∏—Ç—É–¥—ã —Å–ø–µ–∫—Ç—Ä–∞)
        
        Args:
            sensor_data: DataFrame —Å –¥–∞–Ω–Ω—ã–º–∏ —Å–µ–Ω—Å–æ—Ä–æ–≤
            
        Returns:
            DataFrame —Å –∏–∑–≤–ª–µ—á–µ–Ω–Ω—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏
        """
        features = []
        
        for column in sensor_data.columns:
            signal = sensor_data[column].values
            
            # –ë–∞–∑–æ–≤–∞—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
            signal = signal[np.isfinite(signal)]
            if len(signal) == 0:
                self.logger.warning(f"–ù–µ—Ç –≤–∞–ª–∏–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –∫–æ–ª–æ–Ω–∫–∏ {column}")
                continue
            
            try:
                # –ë–∞–∑–æ–≤—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
                mean_val = np.mean(signal)
                std_val = np.std(signal)
                
                # RMS —Å –∑–∞—â–∏—Ç–æ–π –æ—Ç –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏—è
                signal_squared = np.clip(signal**2, 0, 1e12)
                rms_val = np.sqrt(np.mean(signal_squared))
                
                # –ü–∏–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
                peak_val = np.max(np.abs(signal))
                
                # –ö—Ä–µ—Å—Ç-—Ñ–∞–∫—Ç–æ—Ä —Å –∑–∞—â–∏—Ç–æ–π –æ—Ç –¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ–ª—å
                crest_factor = peak_val / rms_val if rms_val > 1e-12 else 0
                crest_factor = np.clip(crest_factor, 0, 100)  # –û–≥—Ä–∞–∏—á–∏–≤–∞–µ–º —Ä–∞–∑—É–º–Ω—ã–º–∏ –ø—Ä–µ–¥–µ–ª–∞–º–∏
                
                # –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –º–æ–º–µ–Ω—Ç—ã —Å –∑–∞—â–∏—Ç–æ–π –æ—Ç –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏—è
                if std_val > 1e-12:
                    normalized_signal = (signal - mean_val) / std_val
                    # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—â–µ–Ω–∏—è overflow
                    normalized_signal = np.clip(normalized_signal, -50, 50)
                    
                    kurtosis_val = np.mean(normalized_signal**4) - 3
                    skewness_val = np.mean(normalized_signal**3)
                    
                    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
                    kurtosis_val = np.clip(kurtosis_val, -100, 100)
                    skewness_val = np.clip(skewness_val, -100, 100)
                else:
                    kurtosis_val = 0
                    skewness_val = 0
                
                # –§–æ—Ä–º–∏—Ä—É–µ–º –±–∞–∑–æ–≤—ã–π —Å–ª–æ–≤–∞—Ä—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
                features_dict = {
                    f'{column}_mean': mean_val,
                    f'{column}_std': std_val,
                    f'{column}_rms': rms_val,
                    f'{column}_peak': peak_val,
                    f'{column}_kurtosis': kurtosis_val,
                    f'{column}_skewness': skewness_val,
                    f'{column}_crest_factor': crest_factor,
                }
                
                # –ß–∞—Å—Ç–æ—Ç–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å –∑–∞—â–∏—Ç–æ–π –æ—Ç –æ—à–∏–±–æ–∫
                try:
                    # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º —Ä–∞–∑–º–µ—Ä FFT –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
                    fft_size = min(len(signal), 2048)
                    fft = np.fft.fft(signal[:fft_size])
                    magnitude = np.abs(fft[:fft_size//2])  # –¢–æ–ª—å–∫–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ —á–∞—Å—Ç–æ—Ç—ã
                    
                    # –§–∏–ª—å—Ç—Ä—É–µ–º infinity –∏ NaN
                    magnitude = magnitude[np.isfinite(magnitude)]
                    
                    if len(magnitude) > 0:
                        freq_peak = np.max(magnitude)
                        freq_mean = np.mean(magnitude)
                        freq_std = np.std(magnitude)
                        
                        # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ —á–∞—Å—Ç–æ—Ç–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
                        dominant_freq_idx = np.argmax(magnitude)
                        dominant_freq_ratio = dominant_freq_idx / len(magnitude)
                        
                        # –≠–Ω–µ—Ä–≥–∏—è –≤ —Ä–∞–∑–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç–Ω—ã—Ö –ø–æ–ª–æ—Å–∞—Ö
                        low_freq_energy = np.sum(magnitude[:len(magnitude)//4])
                        mid_freq_energy = np.sum(magnitude[len(magnitude)//4:3*len(magnitude)//4])
                        high_freq_energy = np.sum(magnitude[3*len(magnitude)//4:])
                        total_energy = low_freq_energy + mid_freq_energy + high_freq_energy
                        
                        if total_energy > 0:
                            low_freq_ratio = low_freq_energy / total_energy
                            high_freq_ratio = high_freq_energy / total_energy
                        else:
                            low_freq_ratio = high_freq_ratio = 0
                        
                        features_dict.update({
                            f'{column}_freq_peak': freq_peak,
                            f'{column}_freq_mean': freq_mean,
                            f'{column}_freq_std': freq_std,
                            f'{column}_dominant_freq_ratio': dominant_freq_ratio,
                            f'{column}_low_freq_ratio': low_freq_ratio,
                            f'{column}_high_freq_ratio': high_freq_ratio,
                        })
                    else:
                        # –ó–∞–ø–æ–ª–Ω—è–µ–º –Ω—É–ª—è–º–∏ –µ—Å–ª–∏ –Ω–µ —É–¥–∞–ª–æ—Å—å –≤—ã—á–∏—Å–ª–∏—Ç—å
                        features_dict.update({
                            f'{column}_freq_peak': 0,
                            f'{column}_freq_mean': 0,
                            f'{column}_freq_std': 0,
                            f'{column}_dominant_freq_ratio': 0,
                            f'{column}_low_freq_ratio': 0,
                            f'{column}_high_freq_ratio': 0,
                        })
                        
                except Exception as freq_error:
                    self.logger.warning(f"–û—à–∏–±–∫–∞ —á–∞—Å—Ç–æ—Ç–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –¥–ª—è {column}: {freq_error}")
                    features_dict.update({
                        f'{column}_freq_peak': 0,
                        f'{column}_freq_mean': 0,
                        f'{column}_freq_std': 0,
                        f'{column}_dominant_freq_ratio': 0,
                        f'{column}_low_freq_ratio': 0,
                        f'{column}_high_freq_ratio': 0,
                    })
                
                features.append(features_dict)
                
            except Exception as e:
                self.logger.warning(f"–û—à–∏–±–∫–∞ –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è {column}: {e}")
                continue
        
        if not features:
            raise ValueError("–ù–µ —É–¥–∞–ª–æ—Å—å –∏–∑–≤–ª–µ—á—å –Ω–∏ –æ–¥–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞")
        
        # –û–±—ä–µ–¥–∏–Ω—è–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤—Å–µ—Ö –∫–∞–Ω–∞–ª–æ–≤
        combined_features = {}
        for feature_dict in features:
            combined_features.update(feature_dict)
        
        # –§–∏–Ω–∞–ª—å–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ infinity –∏ NaN
        for key, value in combined_features.items():
            if not np.isfinite(value):
                combined_features[key] = 0.0
        
        return pd.DataFrame([combined_features])
    
    def create_sliding_windows_improved(self, sensor_data: pd.DataFrame, 
                                      window_size: int = 1000, 
                                      overlap: float = 0.5) -> Tuple[pd.DataFrame, np.ndarray]:
        """
        –°–æ–∑–¥–∞–µ—Ç —Å–∫–æ–ª—å–∑—è—â–∏–µ –æ–∫–Ω–∞ —Å —É–ª—É—á—à–µ–Ω–Ω–æ–π —ç–≤—Ä–∏—Å—Ç–∏–∫–æ–π –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –∞–Ω–æ–º–∞–ª–∏–π
        
        Args:
            sensor_data: DataFrame —Å –¥–∞–Ω–Ω—ã–º–∏ —Å–µ–Ω—Å–æ—Ä–æ–≤
            window_size: —Ä–∞–∑–º–µ—Ä –æ–∫–Ω–∞ –≤ –æ—Ç—Å—á–µ—Ç–∞—Ö
            overlap: –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è –æ–∫–æ–Ω (0-1)
            
        Returns:
            Tuple –∏–∑ DataFrame —Å –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –∏ –º–∞—Å—Å–∏–≤–∞ –º–µ—Ç–æ–∫
        """
        step_size = int(window_size * (1 - overlap))
        windows_features = []
        labels = []
        
        # –í—ã—á–∏—Å–ª—è–µ–º –≥–ª–æ–±–∞–ª—å–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è —ç–≤—Ä–∏—Å—Ç–∏–∫–∏ –∞–Ω–æ–º–∞–ª–∏–π
        global_stats = {}
        for column in sensor_data.columns:
            signal = sensor_data[column].values
            signal = signal[np.isfinite(signal)]
            
            if len(signal) > 0:
                global_stats[column] = {
                    'mean': np.mean(signal),
                    'std': np.std(signal),
                    'rms': np.sqrt(np.mean(np.clip(signal**2, 0, 1e12))),
                    'percentile_95': np.percentile(np.abs(signal), 95),
                    'percentile_99': np.percentile(np.abs(signal), 99),
                    'median': np.median(signal)
                }
        
        self.logger.info(f"–ì–ª–æ–±–∞–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –≤—ã—á–∏—Å–ª–µ–Ω–∞ –¥–ª—è {len(global_stats)} –∫–∞–Ω–∞–ª–æ–≤")
        
        # –°–æ–∑–¥–∞–µ–º –æ–∫–Ω–∞
        total_windows = (len(sensor_data) - window_size) // step_size + 1
        self.logger.info(f"–ë—É–¥–µ—Ç —Å–æ–∑–¥–∞–Ω–æ ~{total_windows} –æ–∫–æ–Ω")
        
        for i, start in enumerate(range(0, len(sensor_data) - window_size + 1, step_size)):
            end = start + window_size
            window_data = sensor_data.iloc[start:end]
            
            try:
                # –ò–∑–≤–ª–µ–∫–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏–∑ –æ–∫–Ω–∞
                window_features = self.extract_features_safe(window_data)
                
                if window_features.empty:
                    continue
                    
                windows_features.append(window_features.iloc[0])
                
                # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∞–Ω–æ–º–∞–ª—å–Ω–æ—Å—Ç—å –æ–∫–Ω–∞ —Å —É–ª—É—á—à–µ–Ω–Ω–æ–π —ç–≤—Ä–∏—Å—Ç–∏–∫–æ–π
                anomaly_score = self._detect_anomaly_improved(window_data, global_stats)
                is_anomaly = 1 if anomaly_score > 0.35 else 0
                labels.append(is_anomaly)
                
                # –ü—Ä–æ–≥—Ä–µ—Å—Å –∫–∞–∂–¥—ã–µ 500 –æ–∫–æ–Ω
                if (i + 1) % 500 == 0:
                    self.logger.info(f"–û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {i + 1} –æ–∫–æ–Ω...")
                
            except Exception as e:
                self.logger.warning(f"–û—à–∏–±–∫–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –æ–∫–Ω–∞ {start}-{end}: {e}")
                continue
        
        if not windows_features:
            raise ValueError("–ù–µ —É–¥–∞–ª–æ—Å—å —Å–æ–∑–¥–∞—Ç—å –Ω–∏ –æ–¥–Ω–æ–≥–æ –≤–∞–ª–∏–¥–Ω–æ–≥–æ –æ–∫–Ω–∞")
        
        # –§–æ—Ä–º–∏—Ä—É–µ–º –∏—Ç–æ–≥–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
        features_df = pd.DataFrame(windows_features)
        labels_array = np.array(labels)
        
        # –§–∏–Ω–∞–ª—å–Ω–∞—è –æ—á–∏—Å—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö –æ—Ç infinity –∏ NaN
        features_df = features_df.replace([np.inf, -np.inf], np.nan)
        features_df = features_df.fillna(0)
        
        # –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Å–æ–∑–¥–∞–Ω–Ω—ã–º –æ–∫–Ω–∞–º
        n_normal = np.sum(labels_array == 0)
        n_anomaly = np.sum(labels_array == 1)
        
        self.logger.info(f"–°–æ–∑–¥–∞–Ω–æ –æ–∫–æ–Ω: {len(features_df)}")
        self.logger.info(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: –ù–æ—Ä–º–∞={n_normal} ({n_normal/len(labels_array)*100:.1f}%), "
                        f"–ê–Ω–æ–º–∞–ª–∏–∏={n_anomaly} ({n_anomaly/len(labels_array)*100:.1f}%)")
        self.logger.info(f"–†–∞–∑–º–µ—Ä –º–∞—Ç—Ä–∏—Ü—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {features_df.shape}")
        
        return features_df, labels_array
    
    def _detect_anomaly_improved(self, window_data: pd.DataFrame, 
                               global_stats: Dict) -> float:
        """
        –£–ª—É—á—à–µ–Ω–Ω–∞—è —ç–≤—Ä–∏—Å—Ç–∏–∫–∞ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –∞–Ω–æ–º–∞–ª–∏–π
        
        –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç:
        - –û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç –≥–ª–æ–±–∞–ª—å–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫
        - –ù–∞–ª–∏—á–∏–µ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ –æ–∫–Ω–µ
        - –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Å–ø–µ–∫—Ç—Ä–∞–ª—å–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
        - –ù–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –º–æ–º–µ–Ω—Ç—ã
        
        Args:
            window_data: –¥–∞–Ω–Ω—ã–µ –æ–∫–Ω–∞
            global_stats: –≥–ª–æ–±–∞–ª—å–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
            
        Returns:
            float: –æ—Ü–µ–Ω–∫–∞ –∞–Ω–æ–º–∞–ª—å–Ω–æ—Å—Ç–∏ (0-1)
        """
        anomaly_indicators = []
        
        for column in window_data.columns:
            if column not in global_stats:
                continue
                
            signal = window_data[column].values
            signal = signal[np.isfinite(signal)]
            
            if len(signal) < 10:  # –°–ª–∏—à–∫–æ–º –º–∞–ª–æ –¥–∞–Ω–Ω—ã—Ö
                anomaly_indicators.append(0)
                continue
            
            try:
                # –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –æ–∫–Ω–∞
                window_mean = np.mean(signal)
                window_std = np.std(signal)
                window_rms = np.sqrt(np.mean(np.clip(signal**2, 0, 1e12)))
                window_median = np.median(signal)
                
                # –ì–ª–æ–±–∞–ª—å–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
                global_mean = global_stats[column]['mean']
                global_std = global_stats[column]['std']
                global_rms = global_stats[column]['rms']
                global_median = global_stats[column]['median']
                percentile_95 = global_stats[column]['percentile_95']
                percentile_99 = global_stats[column]['percentile_99']
                
                anomaly_score = 0
                
                # 1. –û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
                if global_std > 1e-12:
                    mean_deviation = abs(window_mean - global_mean) / global_std
                    if mean_deviation > 3:
                        anomaly_score += 0.3
                    elif mean_deviation > 2:
                        anomaly_score += 0.2
                    elif mean_deviation > 1.5:
                        anomaly_score += 0.1
                
                # 2. –ò–∑–º–µ–Ω–µ–Ω–∏–µ –≤–∞—Ä–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç–∏
                if global_std > 1e-12:
                    std_ratio = window_std / global_std
                    if std_ratio > 3 or std_ratio < 0.3:
                        anomaly_score += 0.25
                    elif std_ratio > 2 or std_ratio < 0.5:
                        anomaly_score += 0.15
                
                # 3. –í—ã–±—Ä–æ—Å—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –≥–ª–æ–±–∞–ª—å–Ω—ã—Ö –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª–µ–π
                extreme_99_count = np.sum(np.abs(signal) > percentile_99)
                extreme_95_count = np.sum(np.abs(signal) > percentile_95)
                
                extreme_99_ratio = extreme_99_count / len(signal)
                extreme_95_ratio = extreme_95_count / len(signal)
                
                if extreme_99_ratio > 0.10:  # –ë–æ–ª–µ–µ 10% —ç–∫—Å—Ç—Ä–µ–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π (99-–π –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—å)
                    anomaly_score += 0.4
                elif extreme_99_ratio > 0.05:
                    anomaly_score += 0.2
                elif extreme_95_ratio > 0.20:  # –ë–æ–ª–µ–µ 20% –∑–Ω–∞—á–µ–Ω–∏–π –≤—ã—à–µ 95-–≥–æ –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è
                    anomaly_score += 0.15
                
                # 4. RMS –∞–Ω–æ–º–∞–ª–∏–∏
                if global_rms > 1e-12:
                    rms_ratio = window_rms / global_rms
                    if rms_ratio > 3 or rms_ratio < 0.2:
                        anomaly_score += 0.2
                    elif rms_ratio > 2 or rms_ratio < 0.4:
                        anomaly_score += 0.1
                
                # 5. –û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –º–µ–¥–∏–∞–Ω—ã (–±–æ–ª–µ–µ —É—Å—Ç–æ–π—á–∏–≤–æ –∫ –≤—ã–±—Ä–æ—Å–∞–º)
                if global_std > 1e-12:
                    median_deviation = abs(window_median - global_median) / global_std
                    if median_deviation > 2:
                        anomaly_score += 0.15
                
                # 6. –õ–æ–∫–∞–ª—å–Ω—ã–µ –≤—ã–±—Ä–æ—Å—ã –≤ –æ–∫–Ω–µ
                if window_std > 1e-12:
                    local_threshold = window_mean + 3 * window_std
                    local_outliers = np.sum(np.abs(signal) > local_threshold)
                    local_outliers_ratio = local_outliers / len(signal)
                    if local_outliers_ratio > 0.15:
                        anomaly_score += 0.1
                
                # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∏—Ç–æ–≥–æ–≤—É—é –æ—Ü–µ–Ω–∫—É
                anomaly_indicators.append(min(anomaly_score, 1.0))
                
            except Exception as e:
                self.logger.warning(f"–û—à–∏–±–∫–∞ –≤ —ç–≤—Ä–∏—Å—Ç–∏–∫–µ –∞–Ω–æ–º–∞–ª–∏–π –¥–ª—è {column}: {e}")
                anomaly_indicators.append(0)
        
        # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É –∞–Ω–æ–º–∞–ª—å–Ω–æ—Å—Ç–∏ –ø–æ –≤—Å–µ–º –∫–∞–Ω–∞–ª–∞–º
        return np.mean(anomaly_indicators) if anomaly_indicators else 0
    
    def train_model(self, features: pd.DataFrame, labels: np.ndarray, 
                   test_size: float = 0.2) -> Dict[str, Any]:
        """
        –û–±—É—á–∞–µ—Ç –º–æ–¥–µ–ª—å Random Forest —Å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º–∏ –ø—Ä–æ–≤–µ—Ä–∫–∞–º–∏ –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–µ–π
        
        Args:
            features: –º–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
            labels: –º–µ—Ç–∫–∏ –∫–ª–∞—Å—Å–æ–≤
            test_size: —Ä–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
            
        Returns:
            Dict —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –æ–±—É—á–µ–Ω–∏—è –∏ –º–µ—Ç—Ä–∏–∫–∞–º–∏
        """
        try:
            self.logger.info(f"–ù–∞—á–∏–Ω–∞–µ–º –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –Ω–∞ –¥–∞–Ω–Ω—ã—Ö: {features.shape}")
            
            # –ü—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–∞—è –æ—á–∏—Å—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö
            features_clean = features.copy()
            features_clean = features_clean.replace([np.inf, -np.inf], np.nan)
            features_clean = features_clean.fillna(0)
            
            # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤
            unique_labels, label_counts = np.unique(labels, return_counts=True)
            self.logger.info(f"–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {dict(zip(unique_labels, label_counts))}")
            
            # –ï—Å–ª–∏ —Å–ª–∏—à–∫–æ–º –º–∞–ª–æ –∞–Ω–æ–º–∞–ª–∏–π, —Å–æ–∑–¥–∞–µ–º –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è
            if len(unique_labels) < 2 or np.min(label_counts) < 10:
                self.logger.warning("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –ø—Ä–∏–º–µ—Ä–æ–≤ –∞–Ω–æ–º–∞–ª–∏–π –¥–ª—è –æ–±—É—á–µ–Ω–∏—è")
                
                # –°–æ–∑–¥–∞–µ–º –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–µ –∞–Ω–æ–º–∞–ª–∏–∏ (–º–∏–Ω–∏–º—É–º 5% –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞)
                min_anomalies = max(20, len(labels) // 20)
                current_anomalies = np.sum(labels == 1)
                
                if current_anomalies < min_anomalies:
                    additional_anomalies = min_anomalies - current_anomalies
                    normal_indices = np.where(labels == 0)[0]
                    
                    if len(normal_indices) >= additional_anomalies:
                        # –í—ã–±–∏—Ä–∞–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–µ –æ–±—Ä–∞–∑—Ü—ã –∏ –ø–æ–º–µ—á–∞–µ–º –∫–∞–∫ –∞–Ω–æ–º–∞–ª–∏–∏
                        np.random.seed(self.random_state)
                        anomaly_candidates = np.random.choice(normal_indices, 
                                                            additional_anomalies, 
                                                            replace=False)
                        labels[anomaly_candidates] = 1
                        
                        self.logger.info(f"–î–æ–±–∞–≤–ª–µ–Ω–æ {additional_anomalies} –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∞–Ω–æ–º–∞–ª–∏–π")
                        
                        # –û–±–Ω–æ–≤–ª—è–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
                        unique_labels, label_counts = np.unique(labels, return_counts=True)
                        self.logger.info(f"–§–∏–Ω–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {dict(zip(unique_labels, label_counts))}")
            
            # –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
            stratify_labels = labels if len(np.unique(labels)) > 1 else None
            
            X_train, X_test, y_train, y_test = train_test_split(
                features_clean, labels, 
                test_size=test_size, 
                random_state=self.random_state,
                stratify=stratify_labels
            )
            
            self.logger.info(f"–û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞: {X_train.shape}")
            self.logger.info(f"–¢–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞: {X_test.shape}")
            
            # –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
            X_train_scaled = self.scaler.fit_transform(X_train)
            X_test_scaled = self.scaler.transform(X_test)
            
            # –§–∏–Ω–∞–ª—å–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ infinity –ø–æ—Å–ª–µ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
            if not np.all(np.isfinite(X_train_scaled)):
                self.logger.warning("–û–±–Ω–∞—Ä—É–∂–µ–Ω—ã infinity –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ—Å–ª–µ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è, –ø—Ä–∏–º–µ–Ω—è–µ–º –æ—á–∏—Å—Ç–∫—É")
                X_train_scaled = np.nan_to_num(X_train_scaled, nan=0, posinf=0, neginf=0)
                X_test_scaled = np.nan_to_num(X_test_scaled, nan=0, posinf=0, neginf=0)
            
            # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ Random Forest
            self.model = RandomForestClassifier(
                n_estimators=100,          # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤
                max_depth=15,              # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤—å–µ–≤
                min_samples_split=5,       # –ú–∏–Ω–∏–º—É–º –æ–±—Ä–∞–∑—Ü–æ–≤ –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è —É–∑–ª–∞
                min_samples_leaf=2,        # –ú–∏–Ω–∏–º—É–º –æ–±—Ä–∞–∑—Ü–æ–≤ –≤ –ª–∏—Å—Ç–µ
                max_features='sqrt',       # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞ –¥–µ—Ä–µ–≤–æ
                bootstrap=True,            # –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ bootstrap –≤—ã–±–æ—Ä–∫–∏
                class_weight='balanced',   # –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∞ –∫–ª–∞—Å—Å–æ–≤
                random_state=self.random_state,
                n_jobs=-1                  # –ü–∞—Ä–∞–ª–ª–µ–ª—å–Ω–æ–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ
            )
            
            # –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
            self.model.fit(X_train_scaled, y_train)
            self.feature_names = features_clean.columns.tolist()
            self.is_trained = True
            
            # –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
            y_train_pred = self.model.predict(X_train_scaled)
            y_test_pred = self.model.predict(X_test_scaled)
            y_test_proba = self.model.predict_proba(X_test_scaled)[:, 1]
            
            # –ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è
            cv_folds = min(5, len(np.unique(y_train)))  # –û–≥—Ä–∞–∏—á–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–æ–ª–¥–æ–≤
            cv_scores = cross_val_score(
                self.model, X_train_scaled, y_train,
                cv=cv_folds,
                scoring='accuracy'
            )
            
            # –§–æ—Ä–º–∏—Ä—É–µ–º –ø–æ–ª–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
            results = {
                'train_accuracy': accuracy_score(y_train, y_train_pred),
                'test_accuracy': accuracy_score(y_test, y_test_pred),
                'cv_mean': np.mean(cv_scores),
                'cv_std': np.std(cv_scores),
                'classification_report': classification_report(y_test, y_test_pred),
                'confusion_matrix': confusion_matrix(y_test, y_test_pred),
                'feature_importance': dict(zip(self.feature_names, self.model.feature_importances_)),
                'n_train': len(X_train),
                'n_test': len(X_test),
                'n_features': len(self.feature_names),
                'class_distribution': dict(zip(*np.unique(labels, return_counts=True))),
                'cv_scores': cv_scores
            }
            
            # ROC-AUC —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ –µ—Å—Ç—å –æ–±–∞ –∫–ª–∞—Å—Å–∞ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
            if len(np.unique(y_test)) > 1:
                results['roc_auc'] = roc_auc_score(y_test, y_test_proba)
                self.logger.info(f"ROC-AUC Score: {results['roc_auc']:.4f}")
            else:
                results['roc_auc'] = None
                self.logger.info("ROC-AUC –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω (–æ–¥–∏–Ω –∫–ª–∞—Å—Å –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ)")
            
            self.logger.info("‚úÖ –ú–æ–¥–µ–ª—å —É—Å–ø–µ—à–Ω–æ –æ–±—É—á–µ–Ω–∞!")
            self.logger.info(f"–¢–æ—á–Ω–æ—Å—Ç—å –Ω–∞ —Ç–µ—Å—Ç–µ: {results['test_accuracy']:.4f}")
            self.logger.info(f"–ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è: {results['cv_mean']:.4f} ¬± {results['cv_std']:.4f}")
            
            return results
            
        except Exception as e:
            self.logger.error(f"–û—à–∏–±–∫–∞ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏: {e}")
            raise
    
    def predict_anomaly(self, sensor_data: pd.DataFrame, 
                       window_size: int = 1000) -> Tuple[np.ndarray, np.ndarray]:
        """
        –ü—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç –∞–Ω–æ–º–∞–ª–∏–∏ –≤ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
        
        Args:
            sensor_data: –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ —Å–µ–Ω—Å–æ—Ä–æ–≤
            window_size: —Ä–∞–∑–º–µ—Ä –æ–∫–Ω–∞ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
            
        Returns:
            Tuple –∏–∑ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π
        """
        if not self.is_trained:
            raise ValueError("–ú–æ–¥–µ–ª—å –Ω–µ –æ–±—É—á–µ–Ω–∞. –°–Ω–∞—á–∞–ª–∞ –≤—ã–ø–æ–ª–Ω–∏—Ç–µ train_model()")
        
        # –°–æ–∑–¥–∞–µ–º –æ–∫–Ω–∞ –¥–ª—è –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–±–µ–∑ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π)
        features, _ = self.create_sliding_windows_improved(sensor_data, window_size, overlap=0)
        
        # –ü—Ä–∏–º–µ–Ω—è–µ–º —Ç–æ –∂–µ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ —á—Ç–æ –∏ –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏
        features_scaled = self.scaler.transform(features)
        features_scaled = np.nan_to_num(features_scaled, nan=0, posinf=0, neginf=0)
        
        # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
        predictions = self.model.predict(features_scaled)
        probabilities = self.model.predict_proba(features_scaled)[:, 1]
        
        return predictions, probabilities


def visualize_results_comprehensive(results: Dict[str, Any]) -> None:
    """
    –ö–æ–º–ø–ª–µ–∫—Å–Ω–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏
    
    Args:
        results: —Å–ª–æ–≤–∞—Ä—å —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –æ–±—É—á–µ–Ω–∏—è
    """
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    
    # 1. –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
    cm = results['confusion_matrix']
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0,0],
                cbar_kws={'label': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—Ä–∞–∑—Ü–æ–≤'})
    axes[0,0].set_title('–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫', fontsize=14, fontweight='bold')
    axes[0,0].set_ylabel('–ò—Å—Ç–∏–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è')
    axes[0,0].set_xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è')
    
    # –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–¥–ø–∏—Å–∏ –∫ –∫–ª–∞—Å—Å–∞–º
    axes[0,0].set_xticklabels(['–ù–æ—Ä–º–∞', '–ê–Ω–æ–º–∞–ª–∏—è'])
    axes[0,0].set_yticklabels(['–ù–æ—Ä–º–∞', '–ê–Ω–æ–º–∞–ª–∏—è'])
    
    # 2. –í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (—Ç–æ–ø-15)
    importance_items = list(results['feature_importance'].items())
    importance_items.sort(key=lambda x: x[1], reverse=True)
    top_features = importance_items[:15]
    
    features, importances = zip(*top_features)
    y_pos = np.arange(len(features))
    
    bars = axes[0,1].barh(y_pos, importances, color='skyblue', edgecolor='navy', alpha=0.8)
    axes[0,1].set_yticks(y_pos)
    axes[0,1].set_yticklabels([f.replace('sensor_1_', '') for f in features], fontsize=9)
    axes[0,1].set_xlabel('–í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∞')
    axes[0,1].set_title('–¢–æ–ø-15 –≤–∞–∂–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤', fontsize=14, fontweight='bold')
    axes[0,1].invert_yaxis()
    
    # –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ –±–∞—Ä—ã
    for i, (bar, imp) in enumerate(zip(bars, importances)):
        axes[0,1].text(bar.get_width() + max(importances)*0.01, 
                      bar.get_y() + bar.get_height()/2, 
                      f'{imp:.3f}', ha='left', va='center', fontsize=8)
    
    # 3. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤
    class_dist = results['class_distribution']
    classes = list(class_dist.keys())
    counts = list(class_dist.values())
    class_names = ['–ù–æ—Ä–º–∞' if c == 0 else '–ê–Ω–æ–º–∞–ª–∏—è' for c in classes]
    colors = ['lightblue', 'lightcoral']
    
    wedges, texts, autotexts = axes[0,2].pie(counts, labels=class_names, autopct='%1.1f%%', 
                                            colors=colors, startangle=90, 
                                            explode=(0.05, 0.05) if len(counts) > 1 else None)
    axes[0,2].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö', fontsize=14, fontweight='bold')
    
    # –£–ª—É—á—à–∞–µ–º –≤–Ω–µ—à–Ω–∏–π –≤–∏–¥ pie chart
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')
        autotext.set_fontsize(12)
    
    # 4. –ú–µ—Ç—Ä–∏–∫–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
    axes[1,0].axis('off')
    
    # –§–æ—Ä–º–∏—Ä—É–µ–º —Ç–µ–∫—Å—Ç —Å –º–µ—Ç—Ä–∏–∫–∞–º–∏
    roc_auc_text = f"{results['roc_auc']:.4f}" if results['roc_auc'] is not None else "N/A"
    
    metrics_text = f"""–†–ï–ó–£–õ–¨–¢–ê–¢–´ –ú–û–î–ï–õ–ò:

üéØ –¢–û–ß–ù–û–°–¢–¨:
‚Ä¢ –ù–∞ –æ–±—É—á–µ–Ω–∏–∏: {results['train_accuracy']:.4f}
‚Ä¢ –ù–∞ —Ç–µ—Å—Ç–µ: {results['test_accuracy']:.4f}

üìä –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–´–ï –ú–ï–¢–†–ò–ö–ò:
‚Ä¢ ROC-AUC Score: {roc_auc_text}
‚Ä¢ –ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è: {results['cv_mean']:.4f} ¬± {results['cv_std']:.4f}

üìã –†–ê–ó–ú–ï–†–´ –î–ê–ù–ù–´–•:
‚Ä¢ –û–±—É—á–∞—é—â–∏—Ö –ø—Ä–∏–º–µ—Ä–æ–≤: {results['n_train']:,}
‚Ä¢ –¢–µ—Å—Ç–æ–≤—ã—Ö –ø—Ä–∏–º–µ—Ä–æ–≤: {results['n_test']:,}
‚Ä¢ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {results['n_features']:,}

üè∑Ô∏è –ö–õ–ê–°–°–´:
‚Ä¢ –ù–æ—Ä–º–∞: {class_dist.get(0, 0):,} ({class_dist.get(0, 0)/sum(counts)*100:.1f}%)
‚Ä¢ –ê–Ω–æ–º–∞–ª–∏–∏: {class_dist.get(1, 0):,} ({class_dist.get(1, 0)/sum(counts)*100:.1f}%)

ü§ñ –ú–û–î–ï–õ–¨: Random Forest
üå≤ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤: 100
üéõÔ∏è –ë–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∞ –∫–ª–∞—Å—Å–æ–≤: –í–∫–ª—é—á–µ–Ω–∞
‚ö° –ü–∞—Ä–∞–ª–ª–µ–ª—å–Ω–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞: –í–∫–ª—é—á–µ–Ω–∞"""
    
    axes[1,0].text(0.05, 0.95, metrics_text, transform=axes[1,0].transAxes,
                  fontsize=10, verticalalignment='top', fontfamily='monospace',
                  bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgray", alpha=0.8))
    
    # 5. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    all_importances = list(results['feature_importance'].values())
    n_bins = min(20, len(all_importances)//2)
    
    axes[1,1].hist(all_importances, bins=n_bins, color='lightgreen', alpha=0.7, 
                  edgecolor='darkgreen', linewidth=1)
    axes[1,1].set_xlabel('–í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∞')
    axes[1,1].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤')
    axes[1,1].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤', fontsize=14, fontweight='bold')
    
    # –î–æ–±–∞–≤–ª—è–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –ª–∏–Ω–∏–∏
    mean_importance = np.mean(all_importances)
    median_importance = np.median(all_importances)
    
    axes[1,1].axvline(mean_importance, color='red', linestyle='--', 
                     label=f'–°—Ä–µ–¥–Ω–µ–µ: {mean_importance:.4f}')
    axes[1,1].axvline(median_importance, color='orange', linestyle='--',
                     label=f'–ú–µ–¥–∏–∞–Ω–∞: {median_importance:.4f}')
    axes[1,1].legend()
    
    # 6. –ê–Ω–∞–ª–∏–∑ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ —Ç–∏–ø–∞–º
    axes[1,2].axis('off')
    
    # –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –ø–æ —Ç–∏–ø–∞–º –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
    feature_types = {
        'statistical': [],
        'frequency': [], 
        'shape': []
    }
    
    for feature, importance in importance_items:
        if any(stat in feature for stat in ['mean', 'std', 'rms', 'peak']):
            feature_types['statistical'].append((feature, importance))
        elif 'freq' in feature:
            feature_types['frequency'].append((feature, importance))
        elif any(shape in feature for shape in ['kurtosis', 'skewness', 'crest']):
            feature_types['shape'].append((feature, importance))
    
    # –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Ç–æ–ø –ø—Ä–∏–∑–Ω–∞–∫–∏ –ø–æ —Ç–∏–ø–∞–º
    type_colors = {'statistical': 'üîµ', 'frequency': 'üü¢', 'shape': 'üü†'}
    type_names = {
        'statistical': '–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ', 
        'frequency': '–ß–∞—Å—Ç–æ—Ç–Ω—ã–µ', 
        'shape': '–§–æ—Ä–º-—Ñ–∞–∫—Ç–æ—Ä—ã'
    }
    
    feature_analysis_text = "–ê–ù–ê–õ–ò–ó –ü–†–ò–ó–ù–ê–ö–û–í:\\n\\n"
    
    for ftype, features_list in feature_types.items():
        if features_list:
            features_list.sort(key=lambda x: x[1], reverse=True)
            top_5 = features_list[:5]
            
            # –°—Ä–µ–¥–Ω—è—è –≤–∞–∂–Ω–æ—Å—Ç—å –¥–ª—è —Ç–∏–ø–∞
            avg_importance = np.mean([imp for _, imp in features_list])
            
            feature_analysis_text += f"{type_colors[ftype]} {type_names[ftype]} (Œº={avg_importance:.4f}):\\n"
            for i, (fname, fimp) in enumerate(top_5, 1):
                # –£–ø—Ä–æ—â–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞
                short_name = fname.replace('sensor_1_', '').replace('_', ' ')
                feature_analysis_text += f"   {i}. {short_name}: {fimp:.4f}\\n"
            feature_analysis_text += "\\n"
    
    # –î–æ–±–∞–≤–ª—è–µ–º –æ–±—â–∏–π –∞–Ω–∞–ª–∏–∑
    total_features = sum(len(features_list) for features_list in feature_types.values())
    feature_analysis_text += f"üìä –û–ë–©–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:\\n"
    feature_analysis_text += f"‚Ä¢ –í—Å–µ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {results['n_features']}\\n"
    feature_analysis_text += f"‚Ä¢ –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–æ: {total_features}\\n"
    feature_analysis_text += f"‚Ä¢ –ú–∞–∫—Å. –≤–∞–∂–Ω–æ—Å—Ç—å: {max(all_importances):.4f}\\n"
    feature_analysis_text += f"‚Ä¢ –ú–∏–Ω. –≤–∞–∂–Ω–æ—Å—Ç—å: {min(all_importances):.4f}"
    
    axes[1,2].text(0.05, 0.95, feature_analysis_text, transform=axes[1,2].transAxes,
                  fontsize=9, verticalalignment='top', fontfamily='monospace',
                  bbox=dict(boxstyle="round,pad=0.5", facecolor="lightyellow", alpha=0.8))
    
    plt.tight_layout()
    plt.suptitle('–°–∏—Å—Ç–µ–º–∞ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏ —Ç—Ä—É–±–æ–ø—Ä–æ–≤–æ–¥–æ–≤ - –ü–æ–ª–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –Ω–∞ —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö', 
                 fontsize=16, fontweight='bold', y=1.02)
    plt.show()
    
    # –í—ã–≤–æ–¥–∏–º –¥–µ—Ç–∞–ª—å–Ω—ã–π –æ—Ç—á–µ—Ç –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print("\\n" + "="*80)
    print("üìã –î–ï–¢–ê–õ–¨–ù–´–ô –û–¢–ß–ï–¢ –ö–õ–ê–°–°–ò–§–ò–ö–ê–¶–ò–ò")
    print("="*80)
    print(results['classification_report'])
    print("="*80)


def main_complete():
    """
    –ü–û–õ–ù–ê–Ø —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ —Ä–∞–±–æ—Ç—ã —Å–∏—Å—Ç–µ–º—ã –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏
    —Å —Ä–µ–∞–ª—å–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ captured_data.bin
    """
    print("üöÄ –ó–ê–ü–£–°–ö –ü–û–õ–ù–û–ô –°–ò–°–¢–ï–ú–´ –î–ò–ê–ì–ù–û–°–¢–ò–ö–ò –¢–†–£–ë–û–ü–†–û–í–û–î–û–í")
    print("="*70)
    print("–ê–≤—Ç–æ—Ä: –ö–æ–Ω–¥—Ä–∞—à–æ–≤ –î.–í.")
    print("–ù–∞—É—á–Ω—ã–π —Ä—É–∫–æ–≤–æ–¥–∏—Ç–µ–ª—å: –ó–∞—Ä—É–±–∏–Ω –ê.–ì.")
    print("–°–∏—Å—Ç–µ–º–∞: Random Forest + MIMII Dataset")
    print("="*70)
    
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Å–∏—Å—Ç–µ–º—ã
    system = PipelineRealDataSystemFixed(random_state=42)
    
    # –ü—É—Ç—å –∫ –≤–∞—à–µ–º—É —Ñ–∞–π–ª—É –¥–∞–Ω–Ω—ã—Ö
    filepath = r"C:\Users\kondr\OneDrive\Desktop\Coding\Petrol-Pipeline-Defect-Detection-main\Petrol-Pipeline-Defect-Detection-main\captured_data.bin"
    
    try:
        print("\nüîÑ –≠—Ç–∞–ø 1: –ó–∞–≥—Ä—É–∑–∫–∞ —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö...")
        raw_data = system.load_binary_data(filepath)
        
        print("\nüìä –≠—Ç–∞–ø 2: –°—Ç—Ä—É–∫—Ç—É—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö...")
        sensor_data = system.parse_sensor_data(raw_data)
        
        print("\nüîß –≠—Ç–∞–ø 3: –°–æ–∑–¥–∞–Ω–∏–µ –æ–∫–æ–Ω –∏ –∏–∑–≤–ª–µ—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤...")
        features, labels = system.create_sliding_windows_improved(
            sensor_data, window_size=1000, overlap=0.5
        )
        
        print("\nüéì –≠—Ç–∞–ø 4: –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ Random Forest...")
        results = system.train_model(features, labels, test_size=0.2)
        
        print("\n‚úÖ –û–ë–£–ß–ï–ù–ò–ï –ó–ê–í–ï–†–®–ï–ù–û –£–°–ü–ï–®–ù–û!")
        print("="*50)
        print(f"üìä –¢–æ—á–Ω–æ—Å—Ç—å –Ω–∞ —Ç–µ—Å—Ç–µ: {results['test_accuracy']:.4f}")
        
        if results['roc_auc'] is not None:
            print(f"üìä ROC-AUC Score: {results['roc_auc']:.4f}")
        else:
            print("üìä ROC-AUC: –ù–µ–¥–æ—Å—Ç—É–ø–Ω–æ (–Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –ø—Ä–∏–º–µ—Ä–æ–≤ –∞–Ω–æ–º–∞–ª–∏–π –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ)")
            
        print(f"üìä –ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è: {results['cv_mean']:.4f} ¬± {results['cv_std']:.4f}")
        print(f"üìä –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {results['class_distribution']}")
        
        # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
        print("\nüìà –î–ï–¢–ê–õ–¨–ù–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
        print(f"   ‚Ä¢ –û–±—É—á–∞—é—â–∏—Ö –ø—Ä–∏–º–µ—Ä–æ–≤: {results['n_train']:,}")
        print(f"   ‚Ä¢ –¢–µ—Å—Ç–æ–≤—ã—Ö –ø—Ä–∏–º–µ—Ä–æ–≤: {results['n_test']:,}")
        print(f"   ‚Ä¢ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {results['n_features']:,}")
        print(f"   ‚Ä¢ –¢–æ—á–Ω–æ—Å—Ç—å –Ω–∞ –æ–±—É—á–µ–Ω–∏–∏: {results['train_accuracy']:.4f}")
        
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º —Ç–æ–ø-10 –≤–∞–∂–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        importance_items = list(results['feature_importance'].items())
        importance_items.sort(key=lambda x: x[1], reverse=True)
        
        print("\nüîß –¢–û–ü-10 –í–ê–ñ–ù–´–• –ü–†–ò–ó–ù–ê–ö–û–í:")
        for i, (feature, importance) in enumerate(importance_items[:10], 1):
            print(f"   {i:2d}. {feature:30s}: {importance:.4f}")
        
        print("\nüé® –≠—Ç–∞–ø 5: –°–æ–∑–¥–∞–Ω–∏–µ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤...")
        visualize_results_comprehensive(results)
        
        return system, results
        
    except FileNotFoundError:
        print(f"‚ùå –û–®–ò–ë–ö–ê: –§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω!")
        print(f"üìÅ –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –ø—É—Ç—å: {filepath}")
        print("üí° –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ —Ñ–∞–π–ª captured_data.bin –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏")
        return None, None
        
    except Exception as e:
        print(f"‚ùå –û–®–ò–ë–ö–ê: {e}")
        print("üí° –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –ª–æ–≥–∏ –≤—ã—à–µ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –ø–æ–¥—Ä–æ–±–Ω–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏")
        return None, None


def predict_new_data_example(system: PipelineRealDataSystemFixed, 
                           new_data_path: str = None) -> None:
    """
    –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    
    Args:
        system: –æ–±—É—á–µ–Ω–Ω–∞—è —Å–∏—Å—Ç–µ–º–∞ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏
        new_data_path: –ø—É—Ç—å –∫ –Ω–æ–≤—ã–º –¥–∞–Ω–Ω—ã–º (–æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ)
    """
    if not system.is_trained:
        print("‚ùå –ú–æ–¥–µ–ª—å –Ω–µ –æ–±—É—á–µ–Ω–∞!")
        return
    
    print("\nüîÆ –ü–†–ò–ú–ï–† –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–Ø –ù–ê –ù–û–í–´–• –î–ê–ù–ù–´–•")
    print("-" * 45)
    
    if new_data_path and os.path.exists(new_data_path):
        # –ó–∞–≥—Ä—É–∂–∞–µ–º –Ω–æ–≤—ã–µ —Ä–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
        new_raw_data = system.load_binary_data(new_data_path)
        new_sensor_data = system.parse_sensor_data(new_raw_data)
        print(f"üìÅ –ó–∞–≥—Ä—É–∂–µ–Ω—ã –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –∏–∑: {new_data_path}")
    else:
        # –°–æ–∑–¥–∞–µ–º —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏
        print("üìù –°–æ–∑–¥–∞–µ–º —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏...")
        n_samples = 10000
        
        # –ù–æ—Ä–º–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ (–±–∞–∑–æ–≤—ã–π —à—É–º)
        normal_data = np.random.normal(0, 0.003, n_samples)
        
        # –î–æ–±–∞–≤–ª—è–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö —É—á–∞—Å—Ç–∫–æ–≤ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏
        anomaly_starts = [2000, 5000, 8000]
        for start in anomaly_starts:
            end = start + 200
            if end < n_samples:
                # –ò–º–∏—Ç–∏—Ä—É–µ–º –∞–Ω–æ–º–∞–ª–∏—é (—É–≤–µ–ª–∏—á–µ–Ω–Ω–∞—è –∞–º–ø–ª–∏—Ç—É–¥–∞ –∏ —á–∞—Å—Ç–æ—Ç–∞)
                anomaly_signal = np.random.normal(0, 0.01, end-start) + \
                               0.02 * np.sin(np.linspace(0, 20*np.pi, end-start))
                normal_data[start:end] = anomaly_signal
        
        new_sensor_data = pd.DataFrame({'sensor_1': normal_data})
        print(f"üìä –°–æ–∑–¥–∞–Ω–æ {len(new_sensor_data)} —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ—Ç—Å—á–µ—Ç–æ–≤ —Å 3 –∞–Ω–æ–º–∞–ª–∏—è–º–∏")
    
    # –î–µ–ª–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    predictions, probabilities = system.predict_anomaly(new_sensor_data, window_size=1000)
    
    # –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
    n_windows = len(predictions)
    n_anomalies = np.sum(predictions == 1)
    anomaly_percentage = n_anomalies / n_windows * 100
    
    print(f"‚úÖ –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–æ {n_windows} –æ–∫–æ–Ω")
    print(f"üö® –û–±–Ω–∞—Ä—É–∂–µ–Ω–æ –∞–Ω–æ–º–∞–ª–∏–π: {n_anomalies} ({anomaly_percentage:.1f}%)")
    
    if n_anomalies > 0:
        print("\nüîç –î–ï–¢–ê–õ–ò –û–ë–ù–ê–†–£–ñ–ï–ù–ù–´–• –ê–ù–û–ú–ê–õ–ò–ô:")
        anomaly_indices = np.where(predictions == 1)[0]
        
        for i, idx in enumerate(anomaly_indices[:5]):  # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 5
            prob = probabilities[idx]
            window_start = idx * 500  # –ü—Ä–∏ overlap=0.5
            print(f"   {i+1}. –û–∫–Ω–æ {idx:3d} (–æ—Ç—Å—á–µ—Ç—ã {window_start:5d}-{window_start+1000:5d}): "
                  f"–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∞–Ω–æ–º–∞–ª–∏–∏ {prob:.3f}")
        
        if len(anomaly_indices) > 5:
            print(f"   ... –∏ –µ—â–µ {len(anomaly_indices) - 5} –∞–Ω–æ–º–∞–ª–∏–π")
            
        # –ê–Ω–∞–ª–∏–∑ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏
        high_conf_anomalies = np.sum(probabilities[predictions == 1] > 0.8)
        print(f"\nüìà –ê–Ω–∞–ª–∏–∑ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏:")
        print(f"   ‚Ä¢ –í—ã—Å–æ–∫–∞—è —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å (>80%): {high_conf_anomalies}/{n_anomalies} –∞–Ω–æ–º–∞–ª–∏–π")
    else:
        print("‚úÖ –ê–Ω–æ–º–∞–ª–∏–π –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ - —Å–∏—Å—Ç–µ–º–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç –≤ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º —Ä–µ–∂–∏–º–µ")


if __name__ == "__main__":
    # –ó–∞–ø—É—Å–∫ –ø–æ–ª–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã
    print("üéØ –ó–∞–ø—É—Å–∫–∞–µ–º –ø–æ–ª–Ω—É—é —Å–∏—Å—Ç–µ–º—É –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏...")
    system, results = main_complete()
    
    # –ï—Å–ª–∏ –æ–±—É—á–µ–Ω–∏–µ –ø—Ä–æ—à–ª–æ —É—Å–ø–µ—à–Ω–æ, –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –Ω–∞ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    if system and results:
        predict_new_data_example(system)
        
        print("\n" + "="*70)
        print("üéâ –°–ò–°–¢–ï–ú–ê –î–ò–ê–ì–ù–û–°–¢–ò–ö–ò –ì–û–¢–û–í–ê –ö –†–ê–ë–û–¢–ï!")
        print("="*70)
        print("‚úÖ –ú–æ–¥–µ–ª—å –æ–±—É—á–µ–Ω–∞ –∏ –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∞ –Ω–∞ —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö captured_data.bin")
        print("‚úÖ –°–æ–∑–¥–∞–Ω–∞ –ø–æ–ª–Ω–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤") 
        print("‚úÖ –°–∏—Å—Ç–µ–º–∞ –≥–æ—Ç–æ–≤–∞ –∫ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–µ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö")
        print("\nüí° –î–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å –Ω–æ–≤—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏:")
        print("   predictions, probabilities = system.predict_anomaly(new_sensor_data)")
        print("\nüîó –ì–æ—Ç–æ–≤–æ –∫ –∏–Ω—Ç–µ–≥—Ä–∞—Ü–∏–∏ –≤ –ø—Ä–æ–º—ã—à–ª–µ–Ω–Ω—É—é —Å—Ä–µ–¥—É!")
```
